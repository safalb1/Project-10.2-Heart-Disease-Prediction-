```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the data
df = pd.read_csv('dataset.csv')

# Data Analysis
# Get the information about the dataset
df.describe()
```

```python
# find the correlation between all the fields
correlation = df.corr()
correlation
```

```python
# check for the null values
df.isnull().sum()
```

# Data Visualization

```python
# Visualize the number of patients having a heart disease and not having a heart disease
sns.countplot(x=df['target'])
plt.title('Heart Disease Count')
plt.xlabel('Disease (1) / No Disease (0)')
plt.ylabel('Count')
plt.show()
```

```python
# Visualize the age and whether the patient is having disease or not
sns.boxplot(x='target', y='age', data=df)
plt.title('Age vs Heart Disease')
plt.xlabel('Disease (1) / No Disease (0)')
plt.ylabel('Age')
plt.show()
```

```python
# Visualize correlation between all features using a heat map
plt.figure(figsize=(12, 8))
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Standardize the features
from sklearn.preprocessing import StandardScaler


X = df.drop(columns=['target'])
y = df['target']
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
# Divide the dataset into 70:30 ratio

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.3, random_state=42)
```

```python
# Build the model on train set and predict the values on test set
log_reg = LogisticRegression(max_iter = 1000)
log_reg.fit(X_train, y_train)
```

```python
logreg_predictions = log_reg.predict(X_test)
```

```python
# Build the confusion matrix and get the accuracy score

from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

logreg_cm = confusion_matrix(y_test, logreg_predictions)
logreg_accuracy = accuracy_score(y_test, logreg_predictions)

print(f'Logistic Regression Confusion Matrix: \n {logreg_cm}')
print(f'Logistic Regression accuracy Score: {logreg_accuracy * 100} ')
```

```python
# Decision Tree
# Build the model on train set and predict the values on test set

from sklearn.tree import DecisionTreeClassifier, plot_tree

dt = DecisionTreeClassifier(random_state=42)

dt.fit(X_train, y_train)
```

```python
dt_predictions = dt.predict(X_test)
```

```python
# Build the Confusion Matrix and calculate the accuracy

dt_cm = confusion_matrix(y_test, dt_predictions)

dt_accuracy = accuracy_score(y_test, dt_predictions)

print(f'Confusion Matrix of the Decision Tree: \n {dt_cm}')
print(f'Accuracy Score of the Decision Tree is {dt_accuracy * 100}')
```

```python
# Random Forest
# Build the model on the train set and predict it on the test set.
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier()

rf.fit(X_train, y_train)
```

```python
rf_predictions = rf.predict(X_test)

# Confusion Matrix
rf_cm = confusion_matrix(y_test, rf_predictions)

# Accuracy score
rf_accuracy = accuracy_score(y_test, rf_predictions)

print(f'Confusion Matrix of Random Forest: \n{rf_cm}')
print(f'Accuracy Score of the Random Forest: {rf_accuracy * 100}')
```

```python
# Select the Best Model
# Classification report of all Classifiers
print('Logistic Regression Classification Report:\n', classification_report(y_test, logreg_predictions))
print('Decision Tree Classification Report: \n', classification_report(y_test, dt_predictions))
print('Random Forest Classification Report: \n', classification_report(y_test, rf_predictions))
```

```python
# Calculate Recall Precision and F1 score of all the models
logreg_report = classification_report(y_test, logreg_predictions, output_dict = True)
dt_report = classification_report(y_test, dt_predictions, output_dict=True)
rf_report = classification_report(y_test, rf_predictions, output_dict=True)
```
